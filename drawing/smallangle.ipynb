# small angle diffraction

```python
import numpy as np  # numpy is a package to use for arrays
import matplotlib.pyplot as plt # to make plot
import pandas as pd
```

```python
# all three data are loaded
# but you can annotate lines in order to only load one set of data, same as later
pdata = np.loadtxt('pout')
ddata = np.loadtxt('dout')
cdata = pd.read_table('out.q',header=None,index_col=False,nrows =12533)
sampledata = np.loadtxt('sout')
```

```python
pdata.shape
ddata.shape
cdata.shape
```

```python
# if you load the cubic data, then this line is needed to calculate the average
c_av=cdata.groupby(0,axis=0,as_index=False).mean()
```

```python
pintensity = pdata[:,1]
pq = pdata[:,0]
dintensity = ddata[:,1]
dq = ddata[:,0]
cintensity = c_av[1]
cq = c_av[0]
```

```python
figq = plt.figure(figsize=(10,7))
cut2 = 0
plt.vlines(cq,0,cintensity,"green",linewidth=1)
# plt.plot(cq,cintensity,'g.',mfc='none')
plt.semilogy(pq[cut2:],pintensity[cut2:],"red",linewidth=2)
plt.plot(dq[cut2:],dintensity[cut2:],"blue",ls=':',linewidth=2)
plt.xticks(np.arange(0,10.1,1),fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("q ($\sigma^{-1}$)",fontsize=20) # set the x-axis label
plt.ylabel(r'$S(q)$',fontsize=20) # set the y-axis label
plt.xlim(0,10) # set the x-axis range

# save the figure, change the name here
figq.savefig('sq_cylinder', dpi = 300,bbox_inches='tight')
```
