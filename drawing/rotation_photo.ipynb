# This is script to generate a photo for rotation method

# load output x, y, I(x,y)

```python
# put the output file here below, notice to rename it or rename here
data = np.loadtxt('out')
```

```python
data.shape
```

```python
# change the range for x and y values according to the input parameters of C codes
y = np.linspace(0, 300, 301)
x = np.linspace(0, 627, 628)
z = data[:,2]
# FI is an error function, used as a threshold for the intensity. change the number below to set the threshold value.
FI = 1 / (1 + np.exp(-z + 8000))
```

```python
X, Y = np.meshgrid(x, y)
# also, notice to change the two values below according to the x and y values above
Z = z.reshape(301,628)
```

```python
fig = plt.figure(figsize=(15,5))
plt.pcolor(X,Y-301//2.0, Z,cmap=plt.cm.Greys_r,vmax=7700)
# this is the lables for x-axis, also change it according to x,y values above.
# plt.xticks([0,78.5,105,157,209,235.5,314,392.5,418,471,522,549.5,627],[r'$-\pi$',r'$-3\pi/4$',r'$-2\pi/3$',r'$-\pi/2$',r'$-\pi/3$',r'$-\pi/4$',r'$0$',r'$\pi/4$',r'$\pi/3$',r'$\pi/2$',r'$2\pi/3$',r'$3\pi/4$',r'$\pi$'])
# plt.colorbar()

# save the figure, change the name here
fig.savefig('baserotated.png', dpi = 300,bbox_inches='tight')
```

```python
plt.figure(figsize=(16,5))

plt.contourf(X, Y, Z,cmap=plt.cm.Greys)
plt.colorbar()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)
```
