```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Ellipse, Circle
```

# load output x, y, I(x,y)

```python
# put the output file here below
data = np.loadtxt('Sq_qxqy.dat')
data.shape
```

```python
# change the W and H values according to the input parameters of C codes
W = 251
H = 251

x = np.linspace(0, W-1, W)
y = np.linspace(0, H-1, H)
z = data[:,3]
```

```python
X, Y = np.meshgrid(x, y)
Z = z.reshape(W,H)
```

```python
# change w to set the size of image
w = 8
fig = plt.figure(figsize=(w,w))
# change vmax to a proper value, in order to see the dots clearly
# it might require multiple times of tries
plt.pcolor(X-W//2.0, Y-H//2.0, Z,cmap=plt.cm.Greys_r,vmax=17)

# save the figure, change the name here
fig.savefig('outputfigure.png', dpi = 300,bbox_inches='tight')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
```
